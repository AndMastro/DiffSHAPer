```python
from matplotlib import pyplot as plt
import pickle
```

```python
# Plot the sample variance for all data indices in a single plot

distance_folder = "results/hausdorff_distances/"

indices_to_plot = [3, 28]


for idx in indices_to_plot:

    plt.figure(figsize=(10, 6))
    with open(f"{distance_folder}{idx}/hausdorff_distances_original.pkl", "rb") as file:
        distances = pickle.load(file)
        plt.plot(range(1, len(distances)+1), distances, label=f"Data Index {idx}")

    with open(f"{distance_folder}{idx}/hausdorff_distances_perturbation_0.pkl", "rb") as file:
        distances = pickle.load(file)
        plt.plot(range(1, len(distances)+1), distances, label=f"Data Index {idx} Perturbation 0")
    with open(f"{distance_folder}{idx}/hausdorff_distances_perturbation_atom_injection.pkl", "rb") as file:
        distances = pickle.load(file)
        plt.plot(range(1, len(distances)+1), distances, label=f"Data Index {idx} Perturbation 0 Injection")
    
    
    plt.rc('axes', labelsize=14)  # Set font size for axes labels
    plt.rc('xtick', labelsize=12)  # Set font size for x-axis ticks
    plt.rc('ytick', labelsize=12)  # Set font size for y-axis ticks
    plt.title('Hausdorff Distance of Generated Linkers', fontsize=14)
    plt.xlabel('Generation frames', fontsize=14)
    plt.ylabel('Distance from original linker ()', fontsize=14)
    custom_legend_names = ["Original", "Atom removal", "Atom injection"]
    plt.legend(custom_legend_names, loc='upper right', fontsize=12, bbox_to_anchor=(1.27, 1))
    # plt.xticks(ticks=range(1, 31, 5))
    plt.xticks(ticks=list(range(1, len(distances), 5)) + [len(distances)], labels=[1] + list(range(5, len(distances), 5)) + [len(distances)])
    plt.grid(True)
    plt.savefig(f"results/hausdorff_distances/hausdorff_distance_trend_{idx}.pdf", dpi=300, bbox_inches='tight')
    plt.savefig(f"results/hausdorff_distances/hausdorff_distance_trend_{idx}.png", dpi=300, bbox_inches='tight')
    plt.show()
```
